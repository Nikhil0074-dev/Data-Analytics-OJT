### 2. In a letter template given below, write a program to fill in with name and date

```python
letter = "Dear <|NAME|>,\n\nYou are selected!\n\n<|DATE|>"

name = input("Enter your name: ")
date = input("Enter date: ")

letter = letter.replace("<|NAME|>", name)
letter = letter.replace("<|DATE|>", date)

print(letter)
```
