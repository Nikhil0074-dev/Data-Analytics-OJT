### 4..Write a program to find whether a given number is prime or not.

````python
n = int(input("Enter number: "))
is_prime = True
for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        is_prime = False
        break
if n > 1 and is_prime:
    print("Prime")
else:
    print("Not prime")
````

### 4.Write a recursive function to calculate the sum of the first n natural numbers.


````python
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n-1)

print(sum_n(5))  # 15
````

### 7..Write a Python function to remove a given word from a list and strip it at the same time.


````python
def remove_word(lst, word):
    lst[:] = [item.strip() for item in lst if item.strip() != word]

lst = [" apple ", "banana", " apple ", "cherry"]
remove_word(lst, "apple")
print(lst)  
````

### 1.Write a program to read the text from a given file poems.txt and find out whether it contains the word "twinkle".

### 2.The game() function in a program lets a user play a game and returns the score as an integer. You need to read a file Hi-score.txt which is either blank or contains the previous high score. Write a program to update the high score whenever game() breaks the Hi-score.


````python
def game():
    return 150

try:
    with open('Hi-score.txt', 'r') as f:
        high = int(f.read().strip())
except:
    high = 0

score = game()
if score > high:
    with open('Hi-score.txt', 'w') as f:
        f.write(str(score))
    print("New high score:", score)
else:
    print("Current high:", high)
````

### 3.Write a program to generate multiplication tables from 2 to 20 and write them to different files. Place these files in a folder.

### 9.Write a program to find out whether a file is identical to and matches the content of another file.

````python
def files_same(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        return f1.read() == f2.read()

print(files_same('this.txt', 'copy.txt'))  # True
````
