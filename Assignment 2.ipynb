### 1. Write a program to find the greatest of four numbers entered by the user.

### 2. Write a program to find out whether a student is pass or fail, if it requires total 40% and at least 33% in each subject. Assume 3 subjects.

```python
marks = [float(input(f"Enter marks for subject {i+1} (out of 100): ")) for i in range(3)]
total = sum(marks)
avg = total / 3
min_pass = 33
if avg >= 40 and all(m >= min_pass for m in marks):
    print("Pass")
else:
    print("Fail")
```

### 3. A spam comment is defined as containing keywords: 'make a lot of money', 'buy now', 'subscribe this', 'click this'. Write a program to detect.

```python
comment = input("Enter comment: ").lower()
spam_keywords = ["make a lot of money", "buy now", "subscribe this", "click this"]
is_spam = any(keyword in comment for keyword in spam_keywords)
print("Spam:", is_spam)
```

### 6. Calculate grade: 90-100 Ex, 80-90 A, 70-80 B, 60-70 C, 50-60 D, <50 F

```python
marks = float(input("Enter marks (0-100): "))
if 90 <= marks <= 100:
    print("Ex")
elif 80 <= marks < 90:
    print("A")
elif 70 <= marks < 80:
    print("B")
elif 60 <= marks < 70:
    print("C")
elif 50 <= marks < 60:
    print("D")
else:
    print("F")
```
